```python
import torch
import torch.nn as nn
# model= u.phonesage(batch_size=1)
model= u.phonesage_norm(batch_size=1)
# model= u.phonesage_cor(batch_size=1)
# model= u.phonesage_frame(batch_size=1)
model.load_state_dict(torch.load(model_path))
model.eval()
loss_func=nn.BCELoss()
```

```python
def confusion_mat(pred, label):#confusion matrix?
    # i2i
    # i2s: truth: iphone, pred: samsung 
    # s2i: truth: samsung, pred: iphone
    # s2s
    from sklearn.metrics import confusion_matrix
    y_pred=pred.max(axis=1).indices.detach().numpy()
    y_label=torch.FloatTensor(label).max(axis=1).indices.detach().numpy()
    return confusion_matrix(y_label, y_pred)
    # i2i, i2s
    # s2i, s2s
```

```python
'''not augmented train accuracy'''
model.batch_size=len(n_train_label)
_,n_train_pred=model(n_train_set, n_train_curve_label, n_train_dist_mat)

n_train_loss=loss_func(n_train_pred, torch.FloatTensor(n_train_label))
n_train_acc=u.binary_acc(n_train_pred, torch.FloatTensor(n_train_label))
n_train_confusion=confusion_mat(n_train_pred, n_train_label)
n_train_confusion=confusion_mat(n_train_pred, n_train_label)
print("not augmented train set")
print("loss:", n_train_loss)
print("accuracy: ", n_train_acc)
print(n_train_confusion)
```

```python
'''not augmented test accuracy'''
model.batch_size=len(n_test_label)
_,n_test_pred=model(n_test_set, n_test_curve_label, n_test_dist_mat)

n_test_loss=loss_func(n_test_pred, torch.FloatTensor(n_test_label))
n_test_acc=u.binary_acc(n_test_pred, torch.FloatTensor(n_test_label))
n_test_confusion=confusion_mat(n_test_pred, n_test_label)
print("not augmented test set")
print("loss:", n_test_loss)
print("accuracy: ", n_test_acc)
print(n_test_confusion)
```
